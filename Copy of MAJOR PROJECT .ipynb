```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7,metric='euclidean')
#k>1
#k!=even
#k=sqrt(49)=7
#Hence , will decide on 7 nearest neighbors
```

```python
#Different classifiers
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import ExtraTreesClassifier
```

```python
#KNeighborsClassifie
step1=ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,3,4])
    ],remainder='passthrough')
step2=KNeighborsClassifier()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(x,y)
y_pred=pipe.predict(x)
y_pred
print("classification report",classification_report(y,y_pred))
print("accuracy score",accuracy_score(y,y_pred))
print("\nConfusion matrix",confusion_matrix(y,y_pred))
```

```python
#ExtraTreesClassifier
step1=ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,3,4])
    ],remainder='passthrough')
step2=ExtraTreesClassifier()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(x,y)
y_pred=pipe.predict(x)
y_pred
print("classification report",classification_report(y,y_pred))
print("accuracy score",accuracy_score(y,y_pred))
print("\nConfusion matrix",confusion_matrix(y,y_pred))
```

```python
%%writefile app.py
import streamlit as st
import pickle
import numpy as np

df=pickle.load(open('df.pkl','rb'))
pipe=pickle.load(open('pipe.pkl','rb'))
st.title("Fruit Classifier")
fruit_label=st.selectbox("The fruit label",df['fruit_label'].unique(),index=0)
mass=st.selectbox("The fruit weighs",df['mass'].unique(),index=0)
width=st.slider("Width of the fruit",min_value=4.5,max_value=10.0,value=5.0,step=0.01)
height=st.selectbox("Height of the fruit",df['height'].unique(),index=4)
color_score=st.selectbox("Color score of the fruit",df['color_score'].unique(),index=0)
if st.button("PREDICT FRUIT"):
    query=np.array([[fruit_label,mass,width,height,color_score]])
    op=pipe.predict(query)
    st.subheader("This fruit is " + str(op[0]))
```
